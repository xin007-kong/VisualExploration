```python
import networkx as nx

# 创建有向图
G = nx.DiGraph()

# 添加节点
G.add_nodes_from(df['head'])
G.add_nodes_from(df['tail'])

# 添加边
for i in range(len(df)):
    head = df.loc[i, 'head']
    tail = df.loc[i, 'tail']
    relation = df.loc[i, 'relation']
    G.add_edge(head, tail, relation=relation)

# 打印节点数和边数
print('Number of nodes:', G.number_of_nodes())
print('Number of edges:', G.number_of_edges())
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import networkx as nx

# 设置中文字体
font_path = 'C:/Windows/Fonts/simhei.ttf'
font_prop = fm.FontProperties(fname=font_path, size=12)

# 创建有向图
G = nx.DiGraph()

# 添加节点
G.add_nodes_from(df['head'])
G.add_nodes_from(df['tail'])

# 添加边
for i in range(len(df)):
    head = df.loc[i, 'head']
    tail = df.loc[i, 'tail']
    relation = df.loc[i, 'relation']
    G.add_edge(head, tail, relation=relation)

# 设置节点和边的样式
pos = nx.spring_layoutpos = nx.spring_layout(G, k=0.5, iterations=100, seed=42)
node_size = [G.degree(node) * 200 for node in G.nodes()]
edge_length = 2
node_size = [G.degree(node) * 100 for node in G.nodes()]
edge_labels = nx.get_edge_attributes(G, 'relation')

# 绘制图形
nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_family='SimHei')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

# 设置图形标题和坐标轴标签
plt.title('水浒传人物关系图', fontproperties=font_prop)
plt.xlabel('人物', fontproperties=font_prop)
plt.ylabel('人物', fontproperties=font_prop)

# 显示图形
plt.figure(figsize=(10, 10))
plt.axis('off')
plt.show()
```

```python
import pandas as pd

# 读取数据
# data = pd.read_csv('triples.csv')
data = df

# 获取所有节点
nodes = list(data['head'].unique()) + list(data['tail'].unique())
nodes = list(set(nodes))  # 去重
num_nodes = len(nodes)

# 创建一个 num_nodes x num_nodes 的矩阵，初始化为0
adjacency_matrix = [[0 for _ in range(num_nodes)] for _ in range(num_nodes)]
adjacency_list = {node: [] for node in nodes}

# 填充邻接矩阵和邻接表
for _, row in data.iterrows():
    i = nodes.index(row['head'])
    j = nodes.index(row['tail'])
    adjacency_matrix[i][j] = 1
    adjacency_list[row['head']].append(row['tail'])
```

```python
import chardet
import codecs
from rdflib import Graph

# 自动检测文件编码格式
with open('shuihuPerson.nt', 'rb') as f:
    result = chardet.detect(f.read())
encoding = result['encoding']

# 读取 .nt 文件并转换为 UTF-8 编码
with codecs.open('shuihuPerson.nt', 'r', encoding) as f:
    content = f.read()
with codecs.open('shuihuPerson_utf8.nt', 'w', 'utf-8') as f:
    f.write(content)

# 创建一个 Graph 对象
g = Graph()

# 读取 UTF-8 编码的 .nt 文件
g.parse('shuihuPerson_utf8.nt', format='nt')

# 遍历三元组
for s, p, o in g:
    print(s, p, o)
```

```python
# 将 RDF 数据转换为 DataFrame
triples = []
for s, p, o in g:
    triples.append((s, p, o))
df = pd.DataFrame(triples, columns=['subject', 'predicate', 'object'])

# 将 DataFrame 导出到 .csv 文件
df.to_csv('example.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup

# 发送 HTTP 请求并获取 HTML 页面
url = 'https://tianchi-jupyter-sh.oss-cn-shanghai.aliyuncs.com/file/opensearch/documents/36027/shuihu.txt'
response = requests.get(url)
html = response.content

# 解析 HTML 页面并提取文本内容
soup = BeautifulSoup(html, 'html.parser')
text = soup.get_text()

# 将文本保存到 .txt 文件中
with open('shuihu.txt', 'w', encoding='utf-8') as f:
    f.write(text)
```
